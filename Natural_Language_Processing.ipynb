# Kleene Closure and Star

```python
if re.match("New(s)*","New"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("New(s)*","News"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("New(s)*","Newssssssssssssssss"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("New(s)+","New"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("New(s)+","News"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("New(s)+","Newsssssssssssss"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("N(e)*w(s)*","Neeeeeeeeeew"):
  print("Match")
else:
  print("Not match")
```

```python
if re.match("N(e)+w(s)*","Neeeeeeeeeew"):
  print("Match")
else:
  print("Not match")
```



# Normalization

```python
# Tokenization
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

sen = "This is a sample sentence .It demonstrates sentence tokenization."
sentences = sent_tokenize(sen)
print(sentences)
```

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

sen = "This is a sample sentence .It demonstrates sentence tokenization."

print(word_tokenize(sen))
```

```python
import nltk
from nltk.tokenize import WhitespaceTokenizer
nltk.download('punkt')

sen = "This is a sample sentence .It demonstrates sentence tokenization."

print(WhitespaceTokenizer().tokenize(sen))
```

```python
import nltk
from nltk.tokenize import wordpunct_tokenize
nltk.download('punkt')

sen = "This is a sample sentence .It demonstrates sentence tokenization."

print(wordpunct_tokenize(sen))
```

# Frequency Distribution

```python
from nltk.probability import FreqDist
f = FreqDist(word_tokenize(sen))
print(f)
```

```python
f.most_common()
```

```python
f.most_common(2)
```

```python
import matplotlib.pyplot as plt
f.plot(30,cumulative=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
f.plot(30,cumulative=True)
plt.show()
```
